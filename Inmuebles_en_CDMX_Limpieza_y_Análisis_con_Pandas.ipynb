##**Demandas de DEV**

### **Crear columnas numéricas**

#### Valor mensual: Gastos mensuales de cada propiedad (alquiler + condominio)

```python
datos = pd.read_csv(url, sep=';')
datos.shape
```

```python
datos['valor_mensual'] = datos['Valor'] + datos['Condominio']
datos.head()
```

#### Valor anual: gastos anuales de propiedad (impuesto + 12 meses de alquiles + condominio)

```python
datos['valor_anual'] = datos['valor_mensual'] *12 + datos['Impuesto']
datos.head()
```

###**Columnas categóricas**

#### **Descripción**: Contiene resumen de la información de las propiedades que se mostrarán en el sitio web.

```python
datos['Descripcion'] = datos['Tipo'] + ' en la colonia ' + datos['Colonia'] + \
                        ' con ' +datos['Habitaciones'].astype(str) + ' habitacion(es) y ' + \
                        datos['Garages'].astype(str) + ' plazas de estacionamiento.'
datos.head()
```

```python
## Revisión de columna
datos.iloc[500].Descripcion
```

####Tiene_suite

```python
datos['tiene_suite'] = datos['Suites'].apply(lambda x: 'Sí' if x > 0 else 'No')
datos.head()
```
